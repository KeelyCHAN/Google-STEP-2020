```python
class mySNS:
    
    def __init__(self, links, nicknames, IDs, me):
        self.links = links
        self.nicknames = nicknames
        self.IDs = IDs
        self.myID = nicknames[me]
        pass

    
    # BFS algorithm for searching a people
    def searchPeople(self, name):
        
        queue = deque()
        queue.append(self.myID)
        
        distance = {}
        distance[self.myID] = 0
        visited_ID = set()
        
        while queue:
            # FIFO
            current_people = queue.popleft()   
            
            # search the neighbours of current people
            for ID in self.links[current_people]: 
                if ID not in visited_ID:
                    visited_ID.add(ID)
                    queue.append(ID)
                    distance[ID] = distance[current_people] + 1
                    
        return distance[self.nicknames[name]]
        pass
    
    
    # recursion part for DFS algorithm
    def recursion(self, currentID, name):
        
        if self.IDs[currentID] == name:
            return True
        
        for ID in self.links[currentID]:
            result = self.recursion(ID, name)
            if result:
                return True
        return False
        pass
    
    
    # DFS algorithm for searching a people
    def searchPeopleDFS(self, name):
        return self.recursionDFS(self.myID, name)
        pass
        
    
    # Find all paired people
    def searchPairedPeople(self):
        
        pairedFriends = []
        
        # find paired ID
        for rootID in self.links:
            for neighbourID in self.links[rootID]:
                if neighbourID in self.links.keys() and rootID in self.links[neighbourID]:
                    pair = list(sorted([IDs[rootID], IDs[neighbourID]])) 
                    if pair not in pairedFriends:
                        pairedFriends.append(pair)
        
        return pairedFriends
        pass
    
    
    # BFS algorithm for finding the furthest people
    def searchFurthestPeople(self):
        
        queue = deque()
        queue.append(self.myID)
        
        distance = {}
        distance[self.myID] = 0
        visited_ID = set()
        
        while queue:
            # FIFO
            current_people = queue.popleft()   
            
            # search the neighbours of current people
            for ID in self.links[current_people]: 
                if ID not in visited_ID:
                    visited_ID.add(ID)
                    queue.append(ID)
                    distance[ID] = distance[current_people] + 1
                    
        max_distance = max(list(distance.values()))
        furthest_people = []
        for k,v in distance.items():
            if v == max_distance:
                furthest_people.append(self.IDs[k])
         
        return max_distance, furthest_people
        pass
```

```python
## pre-processing ##
read_links = file1.read().splitlines()

links_list = []   # raw list of links 
links = {}   # adjacency list of links

for link in read_links:
    link = link.split('\t')
    links_list.append(link)
    
    if link[0] not in links:
        links[link[0]] = [link[1]]
    else: 
        links[link[0]].append(link[1])
        
print('===== Adjacency list of links =====\n\n', links)
```

```python
## pre-processing ##
read_nicknames = file2.read().splitlines()

nicknames = {}
IDs = {}

for name in read_nicknames:
    name = name.split('\t')
    nicknames[name[1]] = name[0]
    IDs[name[0]] = name[1]
    
print('===== Nicknames to IDs =====\n\n', nicknames)
print('\n===== IDs to Nicknames =====\n\n', IDs)
```

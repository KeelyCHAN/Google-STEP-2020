```python
class myWiki:
    
    def __init__(self, links, pages, IDs):
        # {root_ID: [neighbour_IDs]}
        self.links = links
        # {page: ID}
        self.pages = pages
        # {ID: page}
        self.IDs = IDs
        pass
    
    
    def findShortestPath(self, start, end):
        
        distance = {}
        for key in self.IDs.keys():
            distance[key] = float('inf')
        distance[self.pages[start]] = 0
        
        queue = deque()
        queue.append(self.pages[start])
        visited_ID = set()
        
        while queue:
            # FIFO
            current_page = queue.popleft()   
            
            if current_page in self.links.keys():
                for ID in self.links[current_page]: 
                    if ID not in visited_ID:
                        visited_ID.add(ID)
                        queue.append(ID)
                        distance[ID] = distance[current_page] + 1
                    
        shortest = distance[self.pages[end]]
        if shortest != float('inf'):
            return shortest
        else:
            return None
        pass
    
    
    # Find isolated pages without incoming or outcoming edges
    def findIsolatedPages(self):
    
        ID_set = set(sorted(self.IDs.keys()))
    
        for ID in list(ID_set):
            if ID in self.links and self.links[ID]:
                ID_set.remove(ID)
            
        return ID_set
    pass
```

```python
if __name__ == '__main__':
    
    from collections import deque
    
    path = '/Users/keely/lectures/STEP2020/HW 4/wikipedia_links/'
    
    file1 = open(path + 'links.txt', 'r')
    file2 = open(path + 'pages.txt', 'r')
```

```python
## pre-processing ##
read_links = file1.read().splitlines()

links = {}   # adjacency list of links

for link in read_links:
    link = link.split('\t')
    
    if link[0] not in links:
        links[link[0]] = [link[1]]
    else: 
        links[link[0]].append(link[1])
        
print('===== Adjacency list of links =====\n\n', list(links.items())[:5])
```

```python
## pre-processing ##
read_pages = file2.read().splitlines()

pages = {}
IDs = {}

for page in read_pages:
    page = page.split('\t')
    pages[page[1]] = page[0]
    IDs[page[0]] = page[1]
    
print('===== Pages to IDs =====\n\n', list(pages.items())[:5])
print('\n===== IDs to Pages =====\n\n', list(IDs.items())[:5])
```

```python
## initialization ##
solution = myWiki(links, pages, IDs)
```

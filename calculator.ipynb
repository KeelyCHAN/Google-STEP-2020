```python
class calculator:
    
    def __init__(self):
        pass
    
    # read plus
    def plus(self, index):
        token = {'type': '+'}
        return token, index + 1
        pass
    
    # read minus
    def minus(self, index):
        token = {'type': '-'}
        return token, index + 1
        pass
    
    # read times
    def times(self, index):
        token = {'type': '*'}
        return token, index + 1
        pass

    # read divide
    def divide(self, index):
        token = {'type': '/'}
        return token, index + 1
        pass

    # read left bracket
    def leftBracket(self, index):
        token = {'type': '('}
        return token, index + 1
        pass
    
    # read right bracket
    def rightBracket(self, index):
        token = {'type': ')'}
        return token, index + 1
        pass

    # read number
    def number(self, li, index):
        
        num = 0
        n = len(li)
        
        # read number
        while index < n and li[index].isdigit():
                num = num * 10 + int(li[index])
                index += 1
        
        # Special case: digital point 
        if index < n and li[index] == '.':
            index += 1
            temp = 0.1
            while index < n and li[index].isdigit():
                num += int(li[index]) * temp
                temp = temp / 10
                index += 1
                
        token = {'type': 'number', 'number': num}
        return token, index
    
    # tokenize given list
    def tokenize(self, li):
    
        tokens = []
        index = 0
        n = len(li)
    
        # read character and build token
        while index < n:
            ch = li[index]
        
            if ch.isdigit():
                (token, index) = self.number(li, index)
            elif ch == '+':
                (token, index) = self.plus(index)
            elif ch == '-':
                (token, index) = self.minus(index)
            elif ch == '*':
                (token, index) = self.times(index)
            elif ch == '/':
                (token, index) = self.divide(index)
            elif ch == '(':
                (token, index) = self.leftBracket(index)
            elif ch == ')':
                (token, index) = self.rightBracket(index)
            else:
                print('The character', ch, 'is invalid.')
                sys.exit(0)
            tokens.append(token)
            
        return tokens
        pass
    
    def calculate(self, tokens):
    
        # Insert a '+' token at the beginning (for calculating the first num)
        tokens.insert(0, {'type': '+'}) 
    
        stack = []
        n = len(tokens)
        i = 1
    
        while i < n:
            if tokens[i]['type'] == 'number':
                if tokens[i - 1]['type'] == '+':
                    stack.append(tokens[i]['number'])
                
                elif tokens[i - 1]['type'] == '-':
                    stack.append(-1 * tokens[i]['number'])
                
                elif tokens[i - 1]['type'] == '*':
                    stack.append(stack.pop() * tokens[i]['number'])
                
                elif tokens[i - 1]['type'] == '/':
                    stack.append(stack.pop() / tokens[i]['number'])
                  
                # use recursion to deal with bracket
                elif tokens[i - 1]['type'] == '(':
                    num, j = self.calculate(tokens[i:]) 
                    stack.append(num)
                    i += j - 1
                
                elif tokens[i - 1]['type'] == ')':
                    break
                
                # invalid case
                else:
                    print('Check you syntax!')
                    sys.exit(0)
            i += 1
    
        ans = sum(stack)
        return ans, i
```
